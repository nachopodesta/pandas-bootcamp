## 16. Merge in Python

```python
df1 = pd.DataFrame(
    {
     "key":['K0','K1','K2','K3'],
     "A":['A0','A1','A2','A3'],
     "B":['B0','B1','B2','B3'],
    }
)

df2 = pd.DataFrame(
    {
     "key":['K0','K1','K5','K6'],
     "C":['C0','C1','C2','C3'],
     "D":['D0','D1','D2','D3'],
    }
)
```

```python
df1
```

```python
df2
```

```python
pd.merge(df1,df2,on="key",how='inner')
```

```python
df2
```

```python
pd.merge(df1,df2,how='left',on='key')
```

```python
pd.merge(df1,df2,how='right',on='key')
```

```python
pd.merge(df1,df2,how='outer',on='key',indicator=True)
```
